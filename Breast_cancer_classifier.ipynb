# Model Evalauation


```python
X = df_cancer.drop(['target'], axis= 1)
```

```python
X.head()
```

```python
y = df_cancer['target']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.head()
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm , annot=True)
```

# Improving the model using Data normalization

```python
min_train = X_train.min()
```

```python
range_train = (X_train - min_train).max()
```

```python
X_train_scaled = (X_train - min_train) / range_train
```

```python
sns.scatterplot(x= 'mean area', y ='mean smoothness', hue='target', data= df_cancer)
```

```python
sns.scatterplot(x= X_train_scaled['mean area'], y =X_train_scaled['mean smoothness'], hue='target', data= df_cancer)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test) / range_test
```

```python
svc_model.fit(X_train_scaled , y_train)
```

```python
y_predict_scaled = svc_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_predict_scaled)
cm
```

```python
sns.heatmap(cm , annot=True)
```

```python
print(classification_report(y_test, y_predict_scaled))  # after normalization
```

```python
print(classification_report(y_test, y_predict))  # before normalization
```

# Second method for improving the model using C parameter and gamma parameter

```python
param_grid = {'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC() , param_grid , refit= True, verbose=4)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predictions)
```

```python
sns.heatmap(cm , annot=True)
```

```python
print(classification_report(y_test,grid_predictions))
```
